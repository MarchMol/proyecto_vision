## Objetivos:
1. Desarrollar un sistema de vision por computadora que pueda predecir el posisionamiento de logo en un producto dado
2. Los productos deben ser camisetas y pachones
3. Realizar un modulo para la inserción automatica de logo en una imagen de un producto custom.

## Plan de trabajo:

1. **Recopilacion de datos:**

Obtener imagenes de pachones y camisas del proveedor de productos al mayor Makito.es

2. **Data mining**

Obtener las tuplas de imagen + bounding box a partir de procesamiento de imagenes y filtros de morfologia sobre las plantillas del proveedor.

3. **Modulo de validacion IoU manual (estilo captcha)**

Desarrollar un modulo interactivo que le permita a los desarrolladores iterar sobre el conjunto de validacion para seleccionar bounding boxes sobre imagenes complejas para que el modelo tenga una validacion mas robusta.

4. **Entrenamiento de modelo con Transfer learning**

Entrenar una arquitectura de modelo YOLO con esta informacion, aplicando los principios de transfer learning congelando las capas iniciales.

5. **Desarrollo de modulo personalizado para ejecucion peronalizada.**

Dado el modelo entrenado, desarrollar un modulo personalizado para la ejecucion que permita al usuario ingrear una imgen de producto y un logo sin fono y que se comunique con el modelo para generar la imagen resultante.


# Arquitectura:
Se propone utilizar una arquitectura basada en un modelo preentrenado YOLO. Este modelo ya sabria capturar las generalidades de estructura y posicionamiento de imagenes y tiene la estructura adecuada y un formato para predecir bounding boxes.


```
    # Input: imagen en limpio
    # Output: Bounding Box
```

Realizar un abordaje con Transfer Learning a partir de un modelo YOLO tambien nos permitira desarrollar el proyecto incluso con la restriccion de que tenemos poca data (~100 imagenes por clase) y entrenar unicamente las capas mas profundas para obtener nuestro resultado ideal

## Data:
Ya se tiene un origen determinado para obtener las imagenes, aunque requeriran de pre procesamiento para que esten en el formato esperado. La empresa de ventas al por mayor Makito.es provee imagenes de sus productos con una seccion en verde que indica donde deberia ir posicionado el logo, pues proveen servicios de impresion.

Se propone utilizar mascaras para identificar estas areas en las imagenes que las tengan presentes, y posiblemente aplicar filtros de morfologia como erosion y reduccion de ruido, para obtener extraer las bounding boxes en el formato YOLO txt.



# Validacion:
Dado que se contiene data muy especifica y en coniciones ideales para el entrenamiento, se propone un modulo para generar la data de validacion que consistiria de un programa que permita a usuarios ingresar bounding boxes, tipo las verificacion de captcha. Los usuarios, en este caso, seran los mismos miembros de este equpipo quienes iteraran sobre e conjunto de imagenes de validacion y se aplicara un sistema de IoU para obtener el target de este conjunto de datos.

Asimismo, esta es una oportunidad de utilizar imagenes mas complejas, como camisas manchadas, o con fondo distinto o pachones transparentes; para evaluar y fomentar la generalizacion del modelo.


# Modulo personalizado:
Por ultimo, se presenta la idea adicional del modulo personalizado. Una vez entrenado el modelo que reciba una imagen y de como resultado la bounding box, es posible aplicar un modulo final que obtenga esto y le permita al usuario ingresar una imagen de un producto y un logo y recibir las dos unidas, como si fuese para ponerlo en su tienda en linea.
